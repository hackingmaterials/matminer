This notebook demonstrates use of the code 'retrieve_Citrine.py' in plotting data from various datasets collected by Citrine Informatics at http://citrination.com/, and output it in the form of a matplotlib figure. In this example, we query the Citrine API to retrieve all the data on the semiconductor material PbTe contributed by 'OQMD' (http://oqmd.org/) to Citrine. 

```python
from matminer.data_retrieval.retrieve_Citrine import CitrineDataRetrieval
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
api_key = None     # Set your Citrine API key here. If set as an environment variable 'CITRINE_KEY', set it to 'None'
```

Create an adapter to the Citrine Database.

```python
c = CitrineDataRetrieval(api_key)
df = c.get_dataframe(contributor='OQMD', formula='PbTe')
df
```

In the dataframe above, the search query results in 80 samples/hits from the Citrine API, each with 3 properties - band gap, formation energy, and energy relative to convex hull. Below we show a simple example to plot two columns of the dataframe against each other, in this case, the formation energy vs energy relative to convex hull to see if we can pick out any trends.

Now, to plot the two columns,

```python
df.plot(x='Formation energy', y='Stability (energy relative to convex hull)', kind='scatter')
plt.savefig('output_figs/scatter_plot.png')
Image('output_figs/scatter_plot.png')
```

As the formation increases, the compound becomes more unstable, hence leading to a raise in its energy relative to the convex hull.
